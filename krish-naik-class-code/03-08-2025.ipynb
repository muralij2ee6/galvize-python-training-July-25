```python
import numpy as np
from scipy.stats import skew
import matplotlib.pyplot as plt
import seaborn as sns
```

# SKEWNESS

```python
data_normal = np.random.normal(50, 10, 1000)
```

```python
data_pos = np.random.exponential(2, 1000)
```

```python
skew_normal = skew(data_normal)
print(skew_normal)
skew_pos = skew(data_pos)
print(skew_pos)
```

```python
sns.histplot(data_normal, kde=True, color='blue', label="Normal")
sns.histplot(data_pos, kde=True, color='red', label="Positive skew")
plt.legend()
plt.show()
```

# COORELATION AND COVARIANCE

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df.info()
```

```python
print(df.head())
```

```python
df.describe()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

print(numeric_df.columns)
```

```python
numeric_df.head()
```

```python
corr_matrix = numeric_df.corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap of Titanic Numeric Features")
plt.show()
```

# Bernoulli Distribution

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p = 0.7
plt.bar(['0 (Failure)', '1 (Success)'], [1-p, p], color=['red', 'blue'])
plt.title("Bernoulli Distribution (p=0.7)")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.show()
```

```python
user_clicked_on_ad = bernoulli.rvs(0.3, size=100)
print(user_clicked_on_ad)
```

# Binomial Distribution

```python
from scipy.stats import binom

n, p = 15, 0.2
x = range(n+1)
```

```python
plt.bar(x, binom.pmf(x, n, p))
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability")
plt.show()
```
